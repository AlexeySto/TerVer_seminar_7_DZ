1 )  Даны две  независимые выборки. Не соблюдается условие нормальности

x1  380,420, 290

y1 140,360,200,900

Сделайте вывод по результатам, полученным с помощью функции

```python
from scipy.stats import mannwhitneyu

# Данные выборок
x1 = [380, 420, 290]
y1 = [140, 360, 200, 900]

# Применяем тест Манна-Уитни
statistic, p_value = mannwhitneyu(x1, y1, alternative='two-sided')

# Вывод результатов
print(f"Статистика теста: {statistic}")
print(f"p-значение: {p_value}")

alpha = 0.05
if p_value < alpha:
    print("Отвергаем нулевую гипотезу. Существует статистически значимое различие между выборками.")
else:
    print("Нет оснований отвергать нулевую гипотезу. Различия между выборками статистически не значимы.")
```

2 ) Исследовалось влияние препарата на уровень давления пациентов. Сначала измерялось давление до приема препарата, потом через 10 минут и через 30 минут. Есть ли статистически значимые различия?

1е измерение до приема препарата: 150, 160, 165, 145, 155

2е измерение через 10 минут: 140, 155, 150,  130, 135

3е измерение через 30 минут: 130, 130, 120, 130, 125

```python
from scipy.stats import friedmanchisquare

# Данные по измерениям давления
before = [150, 160, 165, 145, 155]
after_10_min = [140, 155, 150, 130, 135]
after_30_min = [130, 130, 120, 130, 125]

# Применяем критерий Фридмана
statistic, p_value = friedmanchisquare(before, after_10_min, after_30_min)

# Вывод результатов
print(f"Статистика теста: {statistic}")
print(f"p-значение: {p_value}")

alpha = 0.05
if p_value < alpha:
    print("Отвергаем нулевую гипотезу. Существуют статистически значимые различия между измерениями.")
else:
    print("Нет оснований отвергать нулевую гипотезу. Различия между измерениями статистически не значимы.")
```

3 ) Сравните 1 и 2 е измерения, предполагая, что 3го измерения через 30 минут не было.

```python
from scipy.stats import wilcoxon

# Данные по измерениям давления
before = [150, 160, 165, 145, 155]
after_10_min = [140, 155, 150, 130, 135]

# Применяем критерий Уилкоксона
statistic, p_value = wilcoxon(before, after_10_min)

# Вывод результатов
print(f"Статистика теста: {statistic}")
print(f"p-значение: {p_value}")

alpha = 0.05
if p_value < alpha:
    print("Отвергаем нулевую гипотезу. Существуют статистически значимые различия между измерениями до приема препарата и через 10 минут.")
else:
    print("Нет оснований отвергать нулевую гипотезу. Различия между измерениями статистически не значимы.")
```

4) Даны 3 группы  учеников плавания.

В 1 группе время на дистанцию 50 м составляют:

56, 60, 62, 55, 71, 67, 59, 58, 64, 67

Вторая группа : 57, 58, 69, 48, 72, 70, 68, 71, 50, 53

Третья группа: 57, 67, 49, 48, 47, 55, 66, 51, 54

```python
from scipy.stats import kruskal

# Данные о времени на дистанцию 50 м для трех групп
group1 = [56, 60, 62, 55, 71, 67, 59, 58, 64, 67]
group2 = [57, 58, 69, 48, 72, 70, 68, 71, 50, 53]
group3 = [57, 67, 49, 48, 47, 55, 66, 51, 54]

# Применяем критерий Крускала - Уоллиса
statistic, p_value = kruskal(group1, group2, group3)

# Вывод результатов
print(f"Статистика теста: {statistic}")
print(f"p-значение: {p_value}")

alpha = 0.05
if p_value < alpha:
    print("Отвергаем нулевую гипотезу. Есть статистически значимые различия во времени на дистанцию 50 метров между группами.")
else:
    print("Нет оснований отвергать нулевую гипотезу. Нет статистически значимых различий во времени на дистанцию 50 метров между группами.")
```
